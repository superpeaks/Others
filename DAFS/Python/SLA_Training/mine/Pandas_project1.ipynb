```python
# iqr method

def iqr(data,col):

    q1 = data[col].quantile(0.25)
    q3 = data[col].quantile(0.75)
    iqr = q3-q1
    
    lower = q1 - (1.5 * iqr)
    upper = q3 + (1.5 * iqr)

    data[col] = np.where(data[col]<lower,lower,data[col])
    data[col] = np.where(data[col]>upper,upper,data[col])

    sns.boxplot(data[col],orient='h')
```
