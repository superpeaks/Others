```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
def outliers(df,col):
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)

    iqr = q3-q1

    upper_band = q3+(1.5*iqr)
    lower_band = q1-(1.5*iqr)

    print(q1,q3,iqr)

    # capping-replacing all higher values with upper bound (75% percentile)
    df[col] = np.where(df[col]>upper_band,upper_band,df[col])         #Capping
    # flooring -replace all lower values with lower bound (25% percentile)
    df[col] = np.where(df[col]<lower_band,lower_band,df[col])         #FlooringS

    plt.figure(figsize=(15,3))
    sns.boxplot(df[col], orient='h')
    plt.show()
    
```
