## tuple()
A **tuple** in Python is an immutable, ordered sequence of elements. Unlike lists, tuples cannot be modified after creation, making them useful for storing fixed data.

### **Properties of Tuples**
1. **Ordered**: Elements maintain their original order.
2. **Immutable**: Once created, elements cannot be changed, added, or removed.
3. **Can Contain Different Data Types**: Tuples can store integers, strings, floats, and even other tuples.
4. **Allow Duplicate Elements**: Multiple occurrences of the same value are permitted.
5. **Indexing and Slicing**: Supports zero-based indexing and slicing like lists.
6. **Hashable**: Tuples can be used as dictionary keys if they contain only hashable elements.

### **Common Tuple Methods**
Python tuples have fewer methods than lists due to their immutability. Here are some useful ones:
1. **count(value)** ‚Äì Returns the number of times `value` appears in the tuple.
   ```python
   t = (1, 2, 3, 2, 2, 4)
   print(t.count(2))  # Output: 3
   ```
2. **index(value)** ‚Äì Returns the index of the first occurrence of `value`.
   ```python
   t = (10, 20, 30, 40)
   print(t.index(30))  # Output: 2
   ```
3. **len(tuple)** ‚Äì Returns the number of elements.
   ```python
   t = (5, 10, 15)
   print(len(t))  # Output: 3
   ```
4. **tuple(iterable)** ‚Äì Converts an iterable (list, set, string) into a tuple.
   ```python
   lst = [1, 2, 3]
   t = tuple(lst)
   print(t)  # Output: (1, 2, 3)
   ```
5. **max(tuple) & min(tuple)** ‚Äì Returns the maximum or minimum value.
   ```python
   t = (4, 7, 1, 9)
   print(max(t))  # Output: 9
   print(min(t))  # Output: 1
   ```

## set{}
A **set** in Python is an unordered collection of unique elements, meaning it automatically removes duplicates and doesn't maintain any specific order.

### **Properties of Sets**
1. **Unordered**: The elements don‚Äôt follow a sequence.
2. **Unique Elements**: No duplicates are allowed.
3. **Mutable**: You can add or remove elements.
4. **Supports Mathematical Operations**: Such as union, intersection, and difference.
5. **Unindexed**: Does not support indexing or slicing like lists or tuples.
6. **Datatype**: Allows only immutable datatypes like string, int, float, bool etc.. but we cannot use list

### **Common Set Methods**
Python sets come with several built-in methods for performing operations like adding, removing, and manipulating elements efficiently. Here‚Äôs a list of commonly used **set methods**:

### **Adding and Removing Elements**
1. **add(value)** ‚Äì Adds an element to the set.
   ```python
   s = {1, 2, 3}
   s.add(4)
   print(s)  # Output: {1, 2, 3, 4}
   ```

2. **remove(value)** ‚Äì Removes a specific element but raises an error if it‚Äôs not found.
   ```python
   s = {10, 20, 30}
   s.remove(20)
   print(s)  # Output: {10, 30}
   ```

3. **discard(value)** ‚Äì Removes an element but doesn‚Äôt raise an error if it doesn‚Äôt exist.
   ```python
   s = {5, 10, 15}
   s.discard(10)
   print(s)  # Output: {5, 15}
   ```

4. **pop()** ‚Äì Removes and returns an arbitrary element (since sets are unordered). pop with indexing raise a error because of set is unorderd and cannot using indexing.
   ```python
   s = {7, 8, 9}
   s.pop()
   print(s)  # Output: {8, 9} (or another variation)
   ```

   ```python
   s = {1, 2, 3}
   s.clear()
   print(s)  # Output: set()
   ```

6. **union(set2)** ‚Äì Returns a new set containing all unique elements from both sets.
   ```python
   s1 = {1, 2, 3}
   s2 = {3, 4, 5}
   print(s1.union(s2))  # Output: {1, 2, 3, 4, 5}
   ```

7. **intersection(set2)** ‚Äì Returns a set containing elements common to both sets.
   ```python
   s1 = {1, 2, 3}
   s2 = {2, 3, 4}
   print(s1.intersection(s2))  # Output: {2, 3}
   ```

8. **difference(set2)** ‚Äì Returns a set containing elements in the first set but not in the second.
   ```python
   s1 = {1, 2, 3}
   s2 = {2, 3, 4}
   ```

9. **symmetric_difference(set2)** ‚Äì Returns elements that are in either of the sets but not both.
   ```python
   s1 = {1, 2, 3}
   s2 = {2, 3, 4}
   print(s1.symmetric_difference(s2))  # Output: {1, 4}
   ```

12. **update(set2)** ‚Äì Adds all unique elements from another set.

14. **difference_update(set2)** ‚Äì Removes elements found in another set.
    ```python

    s1.difference_update(s2)
    ```

15. **symmetric_difference_update(set2)** ‚Äì Updates a set to contain only differing elements.

#### **üîπ Common List Methods**
| Method | Description | Example |
|--------|-------------|---------|
| `append(x)` | Adds an item at the end | ```python lst = [1, 2] lst.append(3) print(lst) # [1, 2, 3]``` |
| `extend(iterable)` | Adds multiple items at once | ```python lst.extend([4, 5])``` |
| `insert(index, value)` | Inserts an item at a specific position | ```python lst.insert(1, "apple")``` |
| `pop(index)` | Removes and returns an item by index (last by default) | ```python lst.pop() # Removes last``` |
| `remove(value)` | Removes the first occurrence of a value | ```python lst.remove(2)``` |
| `reverse()` | Reverses list order | ```python lst.reverse()``` |
| `sort()` | Sorts elements in ascending order | ```python lst.sort()``` |

#### **üîπ Example Usage**
```python
fruits = ["apple", "banana", "cherry"]
fruits.append("mango")  # Adding
fruits.pop()  # Removing last element
fruits.insert(1, "orange")  # Insert at index 1
print(fruits)  # ['apple', 'orange', 'banana']
```

### **2Ô∏è‚É£ Tuple**
A **tuple** is an ordered, **immutable** collection, meaning its values **cannot** be changed.

#### **üîπ Properties of a Tuple**
- **Ordered** ‚úÖ: Elements maintain their sequence.
- **Immutable** ‚ùå: Cannot be changed after creation.
- **Duplicates Allowed** ‚úÖ: Repeated values are possible.
- **Supports Indexing** ‚úÖ: Access via index positions.

#### **üîπ Common Tuple Methods**
| Method | Description | Example |
|--------|-------------|---------|
| `count(value)` | Counts occurrences of a value | ```python tup = (1, 2, 2, 3) print(tup.count(2)) # Output: 2``` |
| `index(value)` | Finds index of first occurrence | ```python tup.index(3)``` |
| `len()` | Finds length | ```python len(tup)``` |

#### **üîπ Example Usage**
```python
coordinates = (10, 20, 30)  # Fixed values
print(coordinates[0])  # Access first element
print(len(coordinates))  # Length of tuple
```

### **3Ô∏è‚É£ Set**
A **set** is an **unordered** collection of **unique** elements.

#### **üîπ Properties of a Set**
- **Unordered** ‚ùå: No guarantee of sequence.
- **Mutable** ‚úÖ: You can add/remove items.
- **No Duplicates** ‚ùå: Prevents repetition.
- **No Indexing** ‚ùå: Elements can‚Äôt be accessed using indexes.

#### **üîπ Common Set Methods**
| Method | Description | Example |
|--------|-------------|---------|
| `add(value)` | Adds an element | ```python s.add(5)``` |
| `remove(value)` | Removes an element (error if missing) | ```python s.remove(10)``` |
| `discard(value)` | Removes an element (no error if missing) | ```python s.discard(20)``` |
| `union(set2)` | Combines two sets | ```python s1.union(s2)``` |
| `intersection(set2)` | Finds common elements | ```python s1.intersection(s2)``` |
| `difference(set2)` | Finds unique elements in the first set | ```python s1.difference(s2)``` |

#### **üîπ Example Usage**
```python
even_numbers = {2, 4, 6, 8}
odd_numbers = {1, 3, 5, 7}
numbers = even_numbers.union(odd_numbers)  # Combines sets
print(numbers)  # {1, 2, 3, 4, 5, 6, 7, 8}
```

### **4Ô∏è‚É£ Dictionary**
A **dictionary** is a key-value mapping that allows **fast lookups**.

#### **üîπ Properties of a Dictionary**
- **Ordered** ‚úÖ (Python 3.7+)
- **Mutable** ‚úÖ: You can add, modify, or remove values.
- **Unique Keys** ‚úÖ: No duplicate keys, but values can repeat.
- **Indexing by Keys** ‚úÖ: Access elements using keys.

#### **üîπ Common Dictionary Methods**
| Method | Description | Example |
|--------|-------------|---------|
| `keys()` | Returns all keys | ```python d.keys()``` |
| `values()` | Returns all values | ```python d.values()``` |
| `items()` | Returns key-value pairs | ```python d.items()``` |
| `get(key)` | Retrieves a value safely | ```python d.get("name")``` |
| `update({key: value})` | Updates values | ```python d.update({"name": "Alice"})``` |
| `pop(key)` | Removes a key-value pair | ```python d.pop("age")``` |

#### **üîπ Example Usage**
```python
student = {"name": "John", "age": 21}
student["grade"] = "A"  # Adding a new key
student.pop("age")  # Removing a key
print(student)  # {'name': 'John', 'grade': 'A'}
```

## **üìå Real-World Applications**
Here‚Äôs how each of these data structures is used in practical scenarios:

### **üîπ List Applications**
‚úÖ **Use Case:** Managing dynamic collections like shopping carts, task lists, or sequences of data that require frequent updates.

**Example: To-Do List App**
```python
tasks = ["Buy groceries", "Complete project", "Workout"]
tasks.append("Read book")  # Adding new tasks dynamically
tasks.remove("Workout")  # Deleting a completed task
print(tasks)  # ['Buy groceries', 'Complete project', 'Read book']
```

Efficiency matters, especially when handling large data. Below is a comparison based on **time complexity** of key operations.

| Operation         | **List**   | **Tuple**  | **Set**   | **Dictionary** |
|------------------|-----------|-----------|----------|---------------|
| Indexing Lookup  | **O(1)**   | **O(1)**   | ‚ùå N/A    | **O(1)** |

### **üîπ Key Takeaways from Performance Comparison**
- **Lists** are great for storing sequential data, but searching/removal takes longer.

- **Dictionaries** excel in storing structured data with fast lookups.
