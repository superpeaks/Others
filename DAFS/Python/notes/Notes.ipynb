## **📌 Python `if` Statement**
The `if` statement in Python is a **conditional** statement that allows **decision-making** in code. It executes a block of code **only if** a specified condition evaluates to `True`.

### **🔹 Basic Syntax**
```python
if condition:
    # Code executes if the condition is True
```

---

## **🔹 Properties of `if` Statements**
- **Evaluates Boolean Expressions** ✅ (Executes if `True`)
- **Supports Logical Operators** (`and`, `or`, `not`)
- **Can Use `elif` and `else` for Multiple Cases**
- **Uses Indentation to Define Code Blocks**
- **Works with Comparison Operators (`==, !=, <, >, <=, >=`)**

---

## **🔹 Types of `if` Statements**
### **1️⃣ Simple `if` Statement**
Executes the code block **only when the condition is `True`.**
```python
age = 18
if age >= 18:
    print("You are eligible to vote!")
```
✅ **Output:** `"You are eligible to vote!"`

---

### **2️⃣ `if-else` Statement**
Executes **one block if `True`, another if `False`.**
```python
temperature = 15
if temperature > 20:
    print("It's warm outside!")
else:
    print("It's cold outside!")
```
✅ **Output:** `"It's cold outside!"`

---

### **3️⃣ `if-elif-else` Statement**
Handles **multiple conditions** using `elif` (else if).
```python
marks = 85
if marks >= 90:
    print("Grade: A")
elif marks >= 80:
    print("Grade: B")
elif marks >= 70:
    print("Grade: C")
else:
    print("Grade: D")
```
✅ **Output:** `"Grade: B"`

---

### **4️⃣ Nested `if` Statement**
An `if` statement inside another `if`.
```python
num = 10
if num > 0:
    print("Positive number")
    if num % 2 == 0:
        print("Even number")
```
✅ **Output:**  
```
Positive number  
Even number  
```

---

### **🔹 Logical Operators in `if` Statements**
Logical operators help combine multiple conditions.

| **Operator** | **Description** | **Example** |
|-------------|---------------|-------------|
| `and` | Returns `True` if **both** conditions are true | `if age >= 18 and country == "India":` |
| `or` | Returns `True` if **any** condition is true | `if age < 18 or country == "India":` |
| `not` | Inverts a Boolean value | `if not has_permission:` |

---

### **🔹 Short-Hand `if` Statement**
Python allows one-line `if` statements:
```python
x = 5
if x > 0: print("Positive number")
```
✅ **Output:** `"Positive number"`

---

### **🔹 `if` with List and String Membership**
#### **Checking if an Element Exists in a List**
```python
fruits = ["apple", "banana", "mango"]
if "banana" in fruits:
    print("Banana is available!")
```
✅ **Output:** `"Banana is available!"`

#### **Checking if a Character Exists in a String**
```python
word = "Python"
if "y" in word:
    print("Letter 'y' found!")
```
✅ **Output:** `"Letter 'y' found!"`

---

### **🔹 Best Practices for `if` Statements**
✅ Use meaningful variable names.  
✅ Avoid deep nesting for better readability.  
✅ Prefer `elif` over multiple `if`s.  
✅ Use logical operators (`and`, `or`) wisely.
