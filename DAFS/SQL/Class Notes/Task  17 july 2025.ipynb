```python
%%sql
-- Create the 'sales' table
CREATE TABLE sales (
    sale_id INT PRIMARY KEY,
    product VARCHAR(50),
    category VARCHAR(50),
    quantity INT,
    price_per_unit DECIMAL(10,2),
    total_amount DECIMAL(10,2),
    sale_date DATE,
    customer_name VARCHAR(50),
    region VARCHAR(50)
);

-- Insert sample data
INSERT INTO sales (sale_id, product, category, quantity, price_per_unit, total_amount, sale_date, customer_name, region)
VALUES
(1, 'Laptop', 'Electronics', 2, 800, 1600, '2025-07-01', 'Alice', 'North'),
(2, 'Phone', 'Electronics', 5, 500, 2500, '2025-07-02', 'Bob', 'South'),
(3, 'Chair', 'Furniture', 10, 70, 700, '2025-07-03', 'Carol', 'North'),
(4, 'Table', 'Furniture', 3, 150, 450, '2025-07-04', 'Dave', 'East'),
(5, 'Monitor', 'Electronics', 4, 200, 800, '2025-07-05', 'Alice', 'North'),
(6, 'Lamp', 'Furniture', 7, 40, 280, '2025-07-06', 'Eve', 'South'),
(7, 'Phone', 'Electronics', 2, 500, 1000, '2025-07-07', 'Frank', 'East'),
(8, 'Laptop', 'Electronics', 1, 800, 800, '2025-07-08', 'Bob', 'South'),
(9, 'Sofa', 'Furniture', 2, 600, 1200, '2025-07-09', 'Carol', 'West'),
(10, 'Laptop', 'Electronics', 1, 800, 800, '2025-07-10', 'Alice', 'North');
```

```python
%%sql
#1.. Find the product with the highest total sales amount.
SELECT product
FROM sales
GROUP BY product
ORDER BY SUM(total_amount) DESC
LIMIT 1;
```

```python
%%sql
#2.Find the customer with the lowest total sales amount.
SELECT customer_name
FROM sales
ORDER BY total_amount ASC
LIMIT 1;
```

```python
%%sql
#3.Find the region with the highest number of sales
SELECT region AS highest_num_of_sale_region
FROM sales
GROUP BY region
ORDER BY COUNT(*) DESC
LIMIT 1;
```

```python
%%sql
#4.Show customers whose total purchases exceed $2000.
SELECT customer_name
FROM SALES
GROUP BY customer_name
ORDER BY SUM(total_amount) DESC
LIMIT 1;
```

```python
%%sql
#5.List products sold in the 'North' region only.
SELECT DISTINCT product
FROM sales
WHERE region LIKE 'north';
```

```python
%%sql
#6.Find the top 3 products with the highest total quantity sold
SELECT product
FROM sales
GROUP BY product
ORDER BY SUM(quantity) DESC
LIMIT 3;
```

```python
%%sql
#7.List customers who made more than one purchase.
SELECT customer_name
FROM sales
GROUP BY customer_name
HAVING COUNT(*)>1;
```

```python
%%sql
#8.Show the average sale amount per region, sorted from highest to lowest
SELECT region,ROUND(AVG(total_amount),2) AS avg_sal
FROM sales
GROUP BY region
ORDER BY avg_sal DESC;
```

```python
%%sql
#9.Find the cheapest product (based on price per unit).
SELECT product AS cheapest_product
FROM sales
ORDER BY price_per_unit ASC
LIMIT 1;
```

```python
%%sql
#10.Find products sold in more than 1 region.
SELECT product
FROM sales
GROUP BY product
HAVING COUNT(DISTINCT region) >1;
```

```python
%%sql
#1. Regions where the average sale amount is higher than the overall average
SELECT region
FROM sales
GROUP BY region
HAVING AVG(total_amount) > (SELECT AVG(total_amount) FROM sales);
```

```python
%%sql
# 2.Sale(s) made on the most recent sale date
SELECT *
FROM sales
ORDER BY sale_date DESC
LIMIT 1;
```

```python
%%sql
#3.Products never sold to customer 'Bob'
SELECT product
FROM sales
WHERE product NOT IN (SELECT DISTINCT product FROM sales WHERE customer_name LIKE 'bob');
```

```python
%%sql
#4.Sales where total_amount is greater than the average total amount
SELECT *
FROM sales
WHERE total_amount > (SELECT AVG(total_amount) FROM sales);
```

```python
%%sql
#Task 9: Find the grade with the maximum number of students.
SELECT grade
FROM student
GROUP BY grade
HAVING COUNT(*) = (SELECT COUNT(*) FROM student 
                      GROUP BY grade ORDER BY COUNT(*) DESC LIMIT 1);
```

```python
%%sql
CREATE TABLE product (
    product_id INT PRIMARY KEY,
    product_name VARCHAR(50),
    price DECIMAL(10,2)
);
INSERT INTO product (product_id, product_name, price) VALUES
(1, 'Product A', 10.00),
(2, 'Product B', 15.00),
(3, 'Product C', 20.00),
(4, 'Product D', 25.00),
(5, 'Product E', 30.00);
```

```python
%%sql
CREATE TABLE sales1 (
    customer_name VARCHAR(50),
    product_id INT,
    quantity INT,
    sale_date DATE,
    total DECIMAL(10,2),
    FOREIGN KEY (product_id) REFERENCES product(product_id)
);

INSERT INTO sales1 (customer_name, product_id, quantity, sale_date, total) VALUES
('John Doe', 1, 2, '2024-01-01', 20.00),
('Jane Smith', 1, 1, '2024-01-15', 10.00),
('Mike Johnson', 2, 3, '2024-01-05', 45.00),
('Emma Brown', 3, 1, '2024-01-12', 20.00),
('John Doe', 4, 2, '2024-01-07', 50.00),
('Jane Smith', 4, 1, '2024-01-25', 25.00),
('Mike Johnson', 4, 1, '2024-01-25', 25.00),
('Sarah White', 5, 1, '2024-01-30', 30.00);
```

```python
%%sql
#1.I NEED THE PRODUCT WHICH SOLD ON 2024-01-05
SELECT p.product_name
FROM product p JOIN sales1 s ON p.product_id = s.product_id
WHERE s.sale_date = "2024-01-05";
```

```python
%%sql
#2.NEED THE PRODUCT NAME PURCHASED BY JOHN DOE
SELECT p.product_name
FROM product p JOIN sales1 s ON p.product_id = s.product_id
WHERE s.customer_name LIKE "john doe";
```

```python
%%sql
#3.WHICH CUSTOMER HAS PURCHASED HIGHEST PRODUCT?
SELECT customer_name
FROM sales1
GROUP BY customer_name
HAVING SUM(quantity) = (SELECT SUM(quantity) FROM sales1
                        GROUP BY customer_name ORDER BY SUM(quantity) DESC
                        LIMIT 1);
```

```python
%%sql
#4.TOTAL SALES OF PRODUCT B
SELECT SUM(total) AS tot_sal_amt_B
FROM sales1
WHERE product_id = (SELECT product_id FROM product
                    WHERE product_name LIKE "product b");
```

```python
%%sql
#5.TOTAL QUANTITY OF PRODUCT B
SELECT SUM(quantity) AS tot_sal_qty_B
FROM sales1
WHERE product_id = (SELECT product_id FROM product
                    WHERE product_name LIKE "product b");
```

```python
%%sql
#SLA(1).DOCX
-- Drop tables if they already exist
DROP TABLE IF EXISTS tasks;
DROP TABLE IF EXISTS employees;

-- Create employees table
CREATE TABLE employees (
    employee_id INT PRIMARY KEY,
    first_name VARCHAR(50),
    middle_name VARCHAR(50),
    last_name VARCHAR(50),
    status ENUM('active', 'inactive'),
    salary DECIMAL(10,2),
    joining_date DATE
);

-- Insert sample employees
INSERT INTO employees (employee_id, first_name, middle_name, last_name, status, salary, joining_date) VALUES
(1, 'John',    'A.', 'Doe',     'active',   55000.00, '2021-06-15'),
(2, 'Jane',    'B.', 'Smith',   'inactive', 62000.00, '2019-04-22'),
(3, 'Alice',   'C.', 'Johnson', 'active',   72000.00, '2020-11-30'),
(4, 'Bob',     'D.', 'Brown',   'active',   48000.00, '2023-01-10'),
(5, 'Charlie', 'E.', 'Davis',   'inactive', 51000.00, '2022-05-25');

-- Create tasks table
CREATE TABLE tasks (
    task_id INT PRIMARY KEY,
    employee_id INT,
    task_name VARCHAR(100),
    status ENUM('completed', 'pending', 'in-progress'),
    assigned_date DATE,
    completed_date DATE,
    FOREIGN KEY (employee_id) REFERENCES employees(employee_id)
);

-- Insert sample tasks
INSERT INTO tasks (task_id, employee_id, task_name, status, assigned_date, completed_date) VALUES
(101, 1, 'Report Analysis',   'completed',   '2024-01-05', '2024-01-20'),
(102, 2, 'Data Cleanup',      'pending',     '2023-12-10', NULL),
(103, 3, 'Presentation Prep', 'completed',   '2024-05-01', '2024-05-05'),
(104, 4, 'Client Feedback',   'in-progress', '2025-06-01', NULL),
(105, 1, 'Budget Forecast',   'completed',   '2024-10-12', '2024-10-25'),
(106, 5, 'Research Survey',   'pending',     '2025-04-15', NULL);
```

```python
%%sql
# 1.	List all active employees with their full names and salaries
SELECT CONCAT(first_name," ",middle_name," ",last_name) AS full_name,salary
FROM employees
WHERE status LIKE 'active';
```

```python
%%sql
#2.	Fetch the names and task names for employees who have completed tasks. 
SELECT e.first_name,e.middle_name,e.last_name,t.task_name
FROM employees e JOIN tasks t ON e.employee_id = t.employee_id
WHERE t.status LIKE "completed";
```

```python
%%sql
#3.	How many tasks are currently pending?
SELECT COUNT(task_name) AS no_of_pending_task
FROM tasks
WHERE status LIKE "pending";
```

```python
%%sql
#4.	Show all tasks assigned to inactive employees.
SELECT t.task_name
FROM employees e JOIN tasks t ON e.employee_id = t.employee_id
WHERE e.status LIKE "inactive";
```

```python
%%sql
#5.	Find the average salary of employees who joined after 2021.
SELECT ROUND(AVG(salary),2) AS avg_sal
FROM employees
WHERE joining_date > "2021-12-31";
```

```python
%%sql
#6.	List all the employees who have not been assigned any task
SELECT first_name,middle_name,last_name
FROM employees
WHERE employee_id NOT IN (SELECT DISTINCT employee_id FROM tasks WHERE assigned_date IS NOT NULL);
```

```python
%%sql
# 7.Get the full name of employees along with the count of tasks assigned to each.
SELECT e.first_name,e.middle_name,e.last_name, COUNT(*) AS num_of_assigned_tasks
FROM employees e JOIN tasks t ON e.employee_id = t.employee_id
GROUP BY e.first_name,e.middle_name,e.last_name;
```

```python
%%sql
# 8.Find the task assigned in the last 6 months with their status and employee name.
SELECT t.task_name,e.first_name,e.middle_name,e.last_name,e.status AS emp_status
FROM employees e JOIN tasks t ON e.employee_id = t.employee_id
WHERE t.assigned_date >= (CURDATE() - INTERVAL 6 MONTH);
```

```python
%%sql
# 9.	List employees with salary >50000 and at least one completed  task.
SELECT DISTINCT e.first_name,e.middle_name,e.last_name,e.salary
FROM employees e JOIN tasks t ON e.employee_id = t.employee_id
WHERE e.salary > 50000 AND t.status LIKE "completed";
```
