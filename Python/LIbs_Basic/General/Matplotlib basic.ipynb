# Matplotlib Examples

```python
# Importing the Matplotlib library
import matplotlib.pyplot as plt
```

## Basic Plotting

```python
# Creating a simple plot
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.plot(x, y, label='Linear')
```

```python
# Adding labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Linear Plot')
plt.legend()
plt.grid(True)
```

## Scatter Plot

```python
# Creating a scatter plot
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.scatter(x, y, color='blue', marker='o', label='Scatter')
```

```python
# Adding labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.legend()
plt.grid(True)
```

## Bar Plot

```python
# Creating a bar plot
categories = ['A', 'B', 'C', 'D', 'E']
values = [10, 25, 15, 30, 20]
plt.bar(categories, values, color='green', label='Bar')
```

```python
# Adding labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot')
plt.legend()
plt.grid(True)
```

## Histogram

```python
# Creating a histogram
data = [5, 10, 15, 20, 25, 30, 35, 40, 45, 50]
plt.hist(data, bins=5, color='purple', alpha=0.7, label='Histogram')
```

```python
# Adding labels and title
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.legend()
plt.grid(True)
```

## Pie Chart

```python
# Creating a pie chart
labels = ['Apples', 'Bananas', 'Cherries', 'Grapes']
sizes = [30, 25, 20, 15]
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=['red', 'yellow', 'orange', 'purple'])
```

```python
# Adding title
plt.title('Pie Chart')
```

## 1. Creating Basic Plots:

```python
import matplotlib.pyplot as plt
import numpy as np

# Line Plot
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
plt.title('Line Plot')
plt.show()

# Scatter Plot
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x, y)
plt.title('Scatter Plot')
plt.show()

# Bar Plot
categories = ['A', 'B', 'C', 'D']
values = [10, 25, 15, 30]
plt.bar(categories, values)
plt.title('Bar Plot')
plt.show()

# Histogram
data = np.random.randn(1000)
plt.hist(data, bins=20, color='purple', alpha=0.7)
plt.title('Histogram')
plt.show()
```

## 2. Customizing Styles:

```python
import matplotlib.pyplot as plt
import numpy as np

# Set a custom color
plt.plot(x, y, color='orange')

# Set a custom line style
plt.plot(x, y, linestyle='dashed')

# Set a custom marker style
plt.scatter(x, y, marker='o')

# Set a custom color map
plt.imshow(np.random.rand(10, 10), cmap='viridis')
```

## 3. Adding Information:

```python
import matplotlib.pyplot as plt
import numpy as np

# Add a title and labels
plt.plot(x, y)
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()

# Add a legend
plt.plot(x, np.sin(x), label='sin(x)')
plt.plot(x, np.cos(x), label='cos(x)')
plt.legend()
plt.title('Line Plot with Legend')
plt.show()
```

## 4. Subplots and Layouts:

```python
import matplotlib.pyplot as plt
import numpy as np

# Subplots
plt.subplot(2, 2, 1)
plt.plot(x, y)
plt.title('Subplot 1')

plt.subplot(2, 2, 2)
plt.scatter(x, y, color='red')
plt.title('Subplot 2')

# Grid Layout
fig, axs = plt.subplots(2, 2)
axs[0, 0].plot(x, y)
axs[0, 1].scatter(x, y, color='green')
plt.tight_layout()
```

## 5. Annotations and Text:

```python
import matplotlib.pyplot as plt
import numpy as np

plt.plot(x, y)
plt.annotate('Peak', xy=(np.pi/2, 1), xytext=(3, 1.5),
             arrowprops=dict(facecolor='black', arrowstyle='->'))
plt.text(7, -0.5, 'Valley', fontsize=12, color='blue')
plt.title('Annotated Line Plot')
plt.show()
```

## 6. Logarithmic Scales:

```python
import matplotlib.pyplot as plt
import numpy as np

# Logarithmic x-axis
plt.semilogx(x, y)
plt.title('Logarithmic X-axis')
plt.show()

# Logarithmic y-axis
plt.semilogy(x, y)
plt.title('Logarithmic Y-axis')
plt.show()

# Logarithmic both axes
plt.loglog(x, y)
plt.title('Logarithmic Both Axes')
plt.show()
```

## 7. Saving Plots:

```python
import matplotlib.pyplot as plt
import numpy as np

plt.plot(x, y)
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.savefig('line_plot.png')  # Save the plot as an image file
plt.show()
```

## 8. Controlling Plot Limits and Ticks:

```python
import matplotlib.pyplot as plt
import numpy as np

plt.plot(x, y)
plt.title('Line Plot')
plt.xlim(0, 10)  # Set x-axis limits
plt.ylim(-1.5, 1.5)  # Set y-axis limits
plt.xticks(np.arange(0, 11, 2))  # Set custom x-axis ticks
plt.yticks([-1, 0, 1])  # Set custom y-axis ticks
plt.show()
```

## 9. Multiple Plots on the Same Axes:

```python
import matplotlib.pyplot as plt
import numpy as np

plt.plot(x, np.sin(x), label='sin(x)')
plt.plot(x, np.cos(x), label='cos(x)')
plt.title('Multiple Plots on the Same Axes')
plt.legend()
plt.show()
```

## 10. Using LaTeX for Text and Labels:

```python
import matplotlib.pyplot as plt
import numpy as np

plt.plot(x, y)
plt.title(r'Line Plot: $y = \sin(x)$')
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
plt.show()
```
