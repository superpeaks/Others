# Seaborn Basics Commands

## 1. Loading and Exploring Data:

```python
import seaborn as sns

# Load a sample dataset (e.g., 'iris')
data = sns.load_dataset('iris')

# Display basic summary statistics
print(data.head())
print(data.describe())
```

## 2. Creating Basic Plots:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Scatter Plot
sns.scatterplot(x='sepal_length', y='sepal_width', data=data)
plt.title('Scatter Plot')
plt.show()

# Line Plot
sns.lineplot(x='sepal_length', y='sepal_width', data=data)
plt.title('Line Plot')
plt.show()

# Bar Plot
sns.barplot(x='species', y='sepal_length', data=data)
plt.title('Bar Plot')
plt.show()

# Histogram
sns.histplot(data['sepal_length'], kde=True)
plt.title('Histogram')
plt.show()
```

## 3. Customizing Styles:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set a custom color palette
custom_palette = sns.color_palette(['#FF5733', '#C70039', '#900C3F', '#581845'])
sns.set_palette(custom_palette)

# Set a custom style
sns.set_style('whitegrid')
```

## 4. Adding Information:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Add a title and labels
sns.scatterplot(x='sepal_length', y='sepal_width', data=data)
plt.title('Scatter Plot of Sepal Dimensions')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()

# Add a legend
sns.lineplot(x='petal_length', y='petal_width', data=data, label='Petal')
sns.lineplot(x='sepal_length', y='sepal_width', data=data, label='Sepal')
plt.title('Line Plot with Legend')
plt.legend()
plt.show()
```

## 5. Categorical Plots:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Categorical Scatter Plot
sns.stripplot(x='species', y='sepal_length', data=data, jitter=True)
plt.title('Categorical Scatter Plot')
plt.show()

# Categorical Box Plot
sns.boxplot(x='species', y='petal_length', data=data)
plt.title('Categorical Box Plot')
plt.show()

# Categorical Bar Plot
sns.barplot(x='species', y='petal_width', data=data)
plt.title('Categorical Bar Plot')
plt.show()
```

## 6. Pair Plots and Correlation Heatmaps:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Pair Plot
sns.pairplot(data, hue='species')
plt.title('Pair Plot')
plt.show()

# Correlation Heatmap
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

## 7. Regression Plots:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Regression Plot
sns.regplot(x='sepal_length', y='sepal_width', data=data)
plt.title('Regression Plot')
plt.show()

# Residual Plot
sns.residplot(x='sepal_length', y='sepal_width', data=data)
plt.title('Residual Plot')
plt.show()

# Logistic Regression Plot
sns.lmplot(x='sepal_length', y='sepal_width', data=data, logistic=True)
plt.title('Logistic Regression Plot')
plt.show()
```

## 8. Distribution Plots:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Histogram and KDE Plot
sns.histplot(data['sepal_length'], kde=True)
plt.title('Histogram and KDE Plot')
plt.show()

# Box Plot
sns.boxplot(x='species', y='petal_length', data=data)
plt.title('Box Plot')
plt.show()

# Kernel Density Plot
sns.kdeplot(data['sepal_width'], shade=True)
plt.title('Kernel Density Plot')
plt.show()
```

## 9. Facet Grids:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# FacetGrid for multi-plot grid
g = sns.FacetGrid(data, col='species')
g.map(sns.scatterplot, 'sepal_length', 'sepal_width')
plt.subplots_adjust(top=0.8)
g.fig.suptitle('FacetGrid of Scatter Plots')
plt.show()

# PairGrid for customizing subplots
g = sns.PairGrid(data, hue='species')
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
g.map_diag(sns.histplot)
g.add_legend()
plt.suptitle('PairGrid of Customized Subplots')
plt.show()
```

## 10. Color Palettes:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Qualitative Color Palette
sns.set_palette('Set2')
sns.barplot(x='species', y='sepal_length', data=data)
plt.title('Bar Plot with Qualitative Color Palette')
plt.show()

# Sequential Color Palette
sns.set_palette('Blues')
sns.lineplot(x='sepal_length', y='sepal_width', data=data)
plt.title('Line Plot with Sequential Color Palette')
plt.show()

# Diverging Color Palette
sns.set_palette('RdBu_r')
sns.boxplot(x='species', y='petal_length', data=data)
plt.title('Box Plot with Diverging Color Palette')
plt.show()
```

## 11. Removing Spines and Ticks:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Removing Spines and Ticks
sns.scatterplot(x='sepal_length', y='sepal_width', data=data)
sns.despine()  # Remove top and right spines
plt.title('Scatter Plot with Removed Spines and Ticks')
plt.xticks([])  # Remove x-axis ticks
plt.yticks([])  # Remove y-axis ticks
plt.show()
```

## 12. Controlling Figure Size:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Custom Figure Size
plt.figure(figsize=(8, 6))
sns.scatterplot(x='sepal_length', y='sepal_width', data=data)
plt.title('Scatter Plot with Custom Figure Size')
plt.show()
```

## 13. Setting Context:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the context
sns.set_context('paper')  # 'paper', 'notebook', 'talk', 'poster'
sns.scatterplot(x='sepal_length', y='sepal_width', data=data)
plt.title('Scatter Plot with Context Set to Paper')
plt.show()
```

## 14. Annotating Plots:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Annotating Points on a Plot
sns.scatterplot(x='sepal_length', y='sepal_width', data=data)
plt.annotate('Outlier', xy=(4.5, 2.5), xytext=(5, 3),
             arrowprops=dict(facecolor='black', arrowstyle='->'))
plt.title('Annotated Scatter Plot')
plt.show()
```

## 15. Changing Color Transparency:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Changing Color Transparency
sns.scatterplot(x='sepal_length', y='sepal_width', data=data, alpha=0.5)
plt.title('Scatter Plot with Color Transparency')
plt.show()
```

## 16. Changing Plot Aesthetics:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Changing Plot Aesthetics
sns.set(style='whitegrid', rc={'axes.facecolor': '#EAEAF2'})
sns.scatterplot(x='sepal_length', y='sepal_width', data=data, palette='dark')
plt.title('Scatter Plot with Changed Aesthetics')
plt.show()
```

## 17. Adding Titles to Subplots:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Adding Titles to Subplots
g = sns.PairGrid(data, hue='species')
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
g.map_diag(sns.histplot)
g.add_legend()
g.fig.subplots_adjust(top=0.9)
g.fig.suptitle('PairGrid of Subplots with Titles')
plt.show()
```
