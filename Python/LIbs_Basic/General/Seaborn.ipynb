# Seaborn

Seaborn is a powerful and user-friendly data visualization library built on top of Matplotlib, another popular visualization library in Python. Seaborn is designed to simplify the process of creating attractive and informative statistical graphics by providing a higher-level interface, more advanced plot styles, and built-in tools for visualizing complex relationships in data. Its main purpose is to enhance the aesthetics and ease of creating visualizations for exploratory data analysis and statistical presentations.

Key purposes and features of Seaborn include:

1. **Attractive Default Styles:** Seaborn comes with visually appealing default themes and color palettes that improve the aesthetics of plots without requiring extensive customization.

2. **High-Level Interface:** Seaborn provides a high-level interface to create complex plots with fewer lines of code compared to Matplotlib. This makes it easier for users to generate informative visualizations quickly.

3. **Statistical Data Visualization:** Seaborn specializes in statistical plots, such as scatter plots, line plots, bar plots, histograms, kernel density plots, and more. It simplifies the creation of visualizations that display statistical relationships in the data.

4. **Data Exploration:** Seaborn offers functions like pair plots, joint plots, and violin plots that facilitate the exploration of relationships between multiple variables in a dataset.

5. **Categorical Data Visualization:** Seaborn provides specific functions to visualize categorical data, including categorical scatter plots, box plots, and bar plots with confidence intervals.

6. **Matrix Plots:** Seaborn supports the creation of heatmaps and clustermaps to visualize matrices or correlation matrices, which are particularly useful in analyzing relationships between multiple variables.

7. **Time Series Visualization:** Seaborn includes functions for visualizing time series data, such as line plots with date-based axes and aggregating data over time periods.

8. **Statistical Estimation:** Seaborn integrates statistical estimation and inference capabilities, such as adding confidence intervals to plots and performing linear regression and logistic regression analysis.

9. **Easy Customization:** While Seaborn aims to provide attractive default styles, it still allows users to customize plots using Matplotlib's underlying functions when necessary.

10. **Integration with Pandas:** Seaborn can directly work with Pandas DataFrames, making it compatible with the data structures commonly used in data analysis.

11. **Interactive Visualization:** Although Seaborn primarily generates static plots, it can be combined with Jupyter Notebook and other interactive tools to create dynamic and interactive visualizations.

In summary, Seaborn complements Matplotlib by providing a higher-level interface and specialized functions for creating visually appealing statistical visualizations. It is particularly well-suited for data exploration, analysis, and presentation in fields such as data science, machine learning, and statistics.

## 1. Attractive Default Styles:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example 1: Default Style
sns.set(style='whitegrid')
tips = sns.load_dataset('tips')
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title('Box Plot with Seaborn Default Style')
plt.show()

# Example 2: Darkgrid Style
sns.set(style='darkgrid')
sns.scatterplot(x='sepal_length', y='petal_length', data=sns.load_dataset('iris'))
plt.title('Scatter Plot with Darkgrid Style')
plt.show()

# Example 3: White Style
sns.set(style='white')
sns.lineplot(x='size', y='total_bill', data=tips)
plt.title('Line Plot with White Style')
plt.show()
```

## 2. High-Level Interface:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example 1: Scatter Plot with Regression Line
sns.scatterplot(x='sepal_length', y='sepal_width', data=sns.load_dataset('iris'))
plt.title('Scatter Plot with Regression Line')
plt.show()

# Example 2: Pair Plot
sns.pairplot(sns.load_dataset('iris'), hue='species')
plt.title('Pair Plot')
plt.show()

# Example 3: Bar Plot with Confidence Intervals
sns.barplot(x='day', y='total_bill', data=tips, ci='sd')
plt.title('Bar Plot with Confidence Intervals')
plt.show()
```

## 3. Statistical Data Visualization:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example 1: Histogram and KDE Plot
sns.histplot(tips['total_bill'], kde=True)
plt.title('Histogram and KDE Plot')
plt.show()

# Example 2: Box Plot with Categorical Data
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title('Box Plot with Categorical Data')
plt.show()

# Example 3: Kernel Density Plot
sns.kdeplot(tips['total_bill'], shade=True)
plt.title('Kernel Density Plot')
plt.show()
```

## 4. Data Exploration:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example 1: Pair Plot with Hue
sns.pairplot(sns.load_dataset('iris'), hue='species')
plt.title('Pair Plot with Hue')
plt.show()

# Example 2: Joint Plot
sns.jointplot(x='sepal_length', y='sepal_width', data=sns.load_dataset('iris'), kind='kde')
plt.title('Joint Plot')
plt.show()

# Example 3: Violin Plot
sns.violinplot(x='day', y='total_bill', data=tips)
plt.title('Violin Plot')
plt.show()
```

## 5. Categorical Data Visualization:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example 1: Categorical Scatter Plot
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True)
plt.title('Categorical Scatter Plot')
plt.show()

# Example 2: Categorical Box Plot
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title('Categorical Box Plot')
plt.show()

# Example 3: Categorical Bar Plot
sns.barplot(x='day', y='total_bill', data=tips, ci='sd')
plt.title('Categorical Bar Plot')
plt.show()
```

## 6. Matrix Plots:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example 1: Heatmap
correlation_matrix = sns.load_dataset('iris').corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Heatmap')
plt.show()

# Example 2: Clustermap
sns.clustermap(correlation_matrix, cmap='coolwarm')
plt.title('Clustermap')
plt.show()

# Example 3: Pair Grid
g = sns.PairGrid(sns.load_dataset('iris'))
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
g.map_diag(sns.histplot)
plt.suptitle('Pair Grid')
plt.show()
```

## 7. Time Series Visualization:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example 1: Time Series Line Plot
flights = sns.load_dataset('flights')
sns.lineplot(x='year', y='passengers', data=flights)
plt.title('Time Series Line Plot')
plt.show()

# Example 2: Time Series Bar Plot
sns.barplot(x='month', y='passengers', data=flights, ci=None)
plt.title('Time Series Bar Plot')
plt.show()

# Example 3: Time Series Point Plot
sns.pointplot(x='month', y='passengers', data=flights, ci=None)
plt.title('Time Series Point Plot')
plt.show()
```

## 8. Statistical Estimation:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example 1: Regression Plot
sns.regplot(x='sepal_length', y='sepal_width', data=sns.load_dataset('iris'))
plt.title('Regression Plot')
plt.show()

# Example 2: Residual Plot
sns.residplot(x='sepal_length', y='sepal_width', data=sns.load_dataset('iris'))
plt.title('Residual Plot')
plt.show()

# Example 3: Logistic Regression Plot
sns.lmplot(x='sepal_length', y='sepal_width', data=sns.load_dataset('iris'), logistic=True)
plt.title('Logistic Regression Plot')
plt.show()
```

## Easy Customization:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example 1: Custom Color Palette
custom_palette = sns.color_palette(['#FF5733', '#C70039', '#900C3F', '#581845'])
sns.set_palette(custom_palette)
sns.boxplot(x='day', y='total_bill', data=sns.load_dataset('tips'))
plt.title('Custom Color Palette')
plt.show()

# Example 2: Custom Plot Style
sns.set_style('whitegrid')
sns.lineplot(x='size', y='total_bill', data=sns.load_dataset('tips'))
plt.title('Custom Plot Style')
plt.show()

# Example 3: Custom Color and Marker
sns.scatterplot(x='sepal_length', y='sepal_width', data=sns.load_dataset('iris'), color='purple', marker='o')
plt.title('Custom Color and Marker')
plt.show()
```

## Integration with Pandas:

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Example 1: Line Plot with Pandas DataFrame
df = pd.DataFrame({'x': [1, 2, 3, 4, 5], 'y': [2, 4, 6, 8, 10]})
sns.lineplot(x='x', y='y', data=df)
plt.title('Line Plot with Pandas DataFrame')
plt.show()

# Example 2: Bar Plot with Pandas DataFrame
df = pd.DataFrame({'Category': ['A', 'B', 'C', 'D'], 'Value': [10, 25, 15, 30]})
sns.barplot(x='Category', y='Value', data=df)
plt.title('Bar Plot with Pandas DataFrame')
plt.show()

# Example 3: Pair Plot with Pandas DataFrame
df = sns.load_dataset('iris')
sns.pairplot(df, hue='species')
plt.title('Pair Plot with Pandas DataFrame')
plt.show()
```

## 11. Interactive Visualization:

```python
import seaborn as sns
import matplotlib.pyplot as plt
from ipywidgets import interact

# Example 1: Interactive Scatter Plot
@interact(points=(10, 100))
def interactive_scatter(points):
    x = sns.load_dataset('iris')['sepal_length'][:points]
    y = sns.load_dataset('iris')['sepal_width'][:points]
    sns.scatterplot(x=x, y=y, hue=sns.load_dataset('iris')['species'][:points])
    plt.title('Interactive Scatter Plot')
    plt.show()

# Example 2: Interactive Pair Plot
@interact(hue=['species', 'sex'])
def interactive_pair_plot(hue):
    df = sns.load_dataset('iris')
    sns.pairplot(df, hue=hue)
    plt.title('Interactive Pair Plot')
    plt.show()

# Example 3: Interactive KDE Plot
@interact(column=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
def interactive_kde(column):
    sns.kdeplot(data=sns.load_dataset('iris')[column], fill=True)
    plt.title('Interactive KDE Plot')
    plt.show()
```
