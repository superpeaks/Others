### Pivot tables

```python
df.head(2)
```

#### Basic Pivot

```python
df_pivot = df.pivot_table(values=['Area','Production'],index=['Season','State_Name'],columns='Crop_Year')
df_pivot
```

### Total Column(Marigin), fill nan value

```python
df_pivot1 = df.pivot_table(values=['Area','Production'],index=['Season','State_Name'],columns='Crop_Year',fill_value=0 , margins=True, margins_name='Total')
df_pivot1.to_excel('pvot1.xlsx',sheet_name='pivot')
df_pivot1
```

### No null drop

```python
df_pivot3 = df.pivot_table(values=['Area','Production'],index=['Season','State_Name'],columns='Crop_Year',dropna = False,fill_value=0 , margins=True, margins_name='Total')
df_pivot3.to_excel('pivot2.xlsx',sheet_name='pivot')
df_pivot3
```

### multiple aggregation column

```python
df_pivot2 = df.pivot_table(values=['Area','Production'],index=['Season','State_Name'],
                          aggfunc={'Area':[np.mean,np.count_nonzero],'Production':[np.mean,max,min,np.sum,np.std]},margins=False)
df_pivot2.to_excel('pivot.xlsx',sheet_name="pivot")
df_pivot2
```

### Concatenating

```python
df.head(2)
```

```python
df_con1 = pd.DataFrame({'A':['A0', 'A1', 'A2'], 'B':['B0','B1','B2'],'C':['C0','C1','C2'],'key1':['K0','K1','K2'],'key0':['K6','K7','K8']},index=[0,1,2])
df_con2 = pd.DataFrame({'A':['A3','A4','A5'], 'B':['B3','B4','B5'], 'C':['C3','C4','C5'],'key2':['K3','K4','K5'],'key0':['K6','K7','K8']}, index = [3,4,5])

print(df_con1)
print(df_con2)
```

```python
pd.concat([df_con1,df_con2],axis=0) #pd.concat([df_con1,df_con2])
```

```python
pd.concat([df_con1,df_con2],axis=1)
```

```python
pd.concat([df_con1,df_con2],join='inner',keys='key0')
```

```python
pd.concat([df_con1,df_con2],join='inner',keys=['key0','key1'])
```

```python
pd.concat([df_con1,df_con2],join='inner',keys=['key0','key1','key2'])
```

```python
pd.concat([df_con1,df_con2],join='outer',keys=['key0','key2'])
```

```python
pd.concat([df_con1,df_con2],join='outer',keys=['key0','key2'],axis = 1)
```

### Merging

```python
pd.merge(left=df_con1,right=df_con2,how='inner',on=['key0'])
```

```python
pd.merge(left=df_con1,right=df_con2,how='outer',on=['key0'])
```

```python
pd.merge(left=df_con1,right=df_con2,how='left',on=['key0'])
```

```python
pd.merge(left=df_con1,right=df_con2,how='right',on=['key0'])
```
